# Model

```python
X, y = datasets.load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
  X_train, X_test, y_train, y_test =  train_test_split(X, y, random_state=42,
                                                       test_size= 0.3)
```

```python
print(X_train.shape, X_test.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)

y_pred
```

#iNterpret

```python
results = pd.DataFrame(np.c_[y_test, y_pred], columns=['real', 'pred'])

results['residual'] = results['real'] - results['pred']

results.drop('residual', axis=1)[:30].plot()
```

```python
results
```

#Optimizacion

```python
linreg_r2 = linreg.score(X_test, y_test)
linreg_coef = linreg.coef_
linreg_mse = mean_squared_error(y_test, y_pred)

linreg_r2
```

```python
pd.Series(linreg_coef, index=diabetes.feature_names).sort_values(ascending=False).plot.bar()
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha=0.1)

lasso.fit(X_train, y_train)
```

```python
lasso_y_pred = lasso.predict(X_test)
```

#### Lasso evaluation

```python
lasso_r2 = lasso.score(X_test, y_test)
lasso_coef = lasso.coef_

lasso_r2, linreg_r2
```

```python
lasso_coef
```

```python
pd.Series(lasso_coef, index=diabetes.feature_names).sort_values(ascending=False).plot.bar()

plt.show()
```

```python
lasso_mse = mean_squared_error(y_test, lasso_y_pred)
lasso_mse, linreg_mse
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1)

ridge.fit(X_train, y_train)

ridge_y_pred = ridge.predict(X_test)
```

#### Ridge evaluation

```python
ridge_r2 = ridge.score(X_test, y_test)
ridge_coef = ridge.coef_

ridge_r2, linreg_r2
```

```python
ridge_coef
```

```python
pd.Series(ridge_coef, index=diabetes.feature_names).sort_values(ascending=False).plot.bar()

plt.show()
```

#Comparacion modelos

```python
plt.figure(figsize=(12, 7))

plt.plot(diabetes.feature_names, linreg_coef, label='Linear Regression', linestyle= 'dotted', marker='*', color='red')
plt.plot(diabetes.feature_names, lasso_coef, label=f'Lasso Regression{lasso.alpha}', linestyle= 'dotted', marker='v', color='blue')
plt.plot(diabetes.feature_names, ridge_coef, label=f'Ridge Regression{ridge.alpha}', linestyle= 'dotted', marker='^', color='green')

plt.legend(loc='best')
plt.show()
```
